```python

import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.cm as cm 

from aodfunctions.testbed import *
```

```python
ideal_depth = 1
trap_x = 0
trap_rayleigh = 2
trap_rad = 1
trap_xfocalshift = 5
```

```python

X, Y = np.meshgrid(np.linspace(-1, 7, 100),  np.linspace(-2, 2, 100)) 

Z = heuristic_3D_potential([Y,0,X],\
    ideal_depth,trap_x,trap_rayleigh,trap_rad,trap_xfocalshift)


plt.pcolormesh(X, Y, Z, cmap = cm.inferno) 
plt.gca().set_aspect('equal')
plt.show()
```

```python

X, Y = np.meshgrid(np.linspace(-1, 7, 20),  np.linspace(-2, 2, 20)) 
# Directional vectors 
u,v,w = heuristic_3D_force([Y,0,X],\
    ideal_depth,trap_x,trap_rayleigh,trap_rad,trap_xfocalshift)
  
# Plotting Vector Field with QUIVER 
plt.quiver(X, Y, w, u, color='g',scale=10) 
plt.title('Vector Field') 
  
# Setting x, y boundary limits 
# plt.xlim(-7, 7) 
# plt.ylim(-7, 7) 
  
# Show plot with grid 
plt.grid() 
plt.show() 
```
